```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('house_train_raw.csv')
test_df = pd.read_csv('houses_test_raw.csv')
```

```python
#se eliminan las columnas con valores nulos

nan_cols_train = [] 
for col in train_df.columns:
    if train_df[col].isna().sum() != 0:
        nan_cols_train.append(col)

train_df2 = train_df.drop(columns=nan_cols_train)


nan_cols_test = [] 
for col in test_df.columns:
    if test_df[col].isna().sum() != 0:
        nan_cols_test.append(col)

test_df2 = test_df.drop(columns=nan_cols_test)

train_df2 = train_df.drop(columns=nan_cols_test)
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
# dropeamos las columnas no relevantes
X_train = train_df2.drop(['SalePrice', 'Id','PoolArea','Condition1', 'Condition2','RoofStyle','LowQualFinSF', 'TotRmsAbvGrd',
       'Fireplaces', 'PavedDrive', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'MiscVal', 'MoSold','GrLivArea',
       'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'Foundation', 'Heating', 'HeatingQC',
       'CentralAir', 'Electrical', 'YearBuilt', 'YearRemodAdd', 'RoofMatl',
       'ExterQual', 'ExterCond', '1stFlrSF', '2ndFlrSF', 'YrSold'], axis='columns')
y_train = train_df2.SalePrice


```

```python
#columnas regresoras
X_train.columns
```

```python
#mapeo de columnas
cols = X_train.columns
columns_transf = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), cols), remainder='passthrough'
)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_log_error

#creamos el modelo

linear_model = LinearRegression()

linear_pipe = make_pipeline(columns_transf, linear_model)
```

```python
#entrenamiento del modelo
linear_pipe.fit(X_train, y_train)
linear_train_preds = linear_pipe.predict(X_train)
mean_squared_log_error(linear_train_preds, y_train)
```
